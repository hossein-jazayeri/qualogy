```python
import pandas as pd
from matplotlib import pyplot as plt
import datetime
import numpy as np
```

```python
glasses = pd.read_csv('data/glasses.csv')
glasses['DATE'] = pd.to_datetime(glasses['DATE'])
glasses = glasses.set_index('DATE')
glasses = glasses.sort_values('DATE')

glasses_grouped = glasses.groupby(pd.TimeGrouper(freq='Min')).agg({
    'ACC_X': 'mean',
    'ACC_Y': 'mean',
    'ACC_Z': 'mean',
    'GYRO_X': 'mean',
    'GYRO_Y': 'mean',
    'GYRO_Z': 'mean',
    'EOG_L': 'mean',
    'EOG_R': 'mean',
    'EOG_H': 'mean',
    'EOG_V': 'mean',
})
glasses_grouped['eog'] = np.sqrt(
    glasses_grouped['EOG_H'] ** 2 + 
    glasses_grouped['EOG_L'] ** 2 + 
    glasses_grouped['EOG_R'] ** 2 + 
    glasses_grouped['EOG_V'] ** 2
)
# eog = glasses_grouped[['eog']]
eog = glasses_grouped[['EOG_V']]
```

```python
smartwatch = pd.read_csv('data/smartwatch.csv')
heart_rate = smartwatch[smartwatch['source']=='heart_rate']
heart_rate['timestamp'] = pd.to_datetime(heart_rate['timestamp'])
heart_rate = heart_rate.sort_values('timestamp')
heart_rate = heart_rate.set_index('timestamp')

heart_rate['heart_rate'] = heart_rate['values'].map(lambda hr: float(hr.split('\'')[1]))
heart_rate = heart_rate.groupby(pd.TimeGrouper(freq='Min')).agg({
    'heart_rate': 'mean',
})
```

```python
smartwatch_acceleration = smartwatch[smartwatch['source']=='linear_acceleration']

smartwatch_acceleration['timestamp'] = pd.to_datetime(smartwatch_acceleration['timestamp'])
smartwatch_acceleration = smartwatch_acceleration.sort_values('timestamp')
smartwatch_acceleration = smartwatch_acceleration.set_index('timestamp')

smartwatch_acceleration['smartwatch_acceleration'] = smartwatch_acceleration['values'].map(
    lambda hr: np.sqrt(
        float(hr.split('\'')[1]) ** 2 + 
        float(hr.split('\'')[3]) ** 2 + 
        float(hr.split('\'')[5]) ** 2
    )
)
smartwatch_acceleration = smartwatch_acceleration.groupby(pd.TimeGrouper(freq='Min')).agg({
    'smartwatch_acceleration': 'mean',
})
```

```python
smartphone = pd.read_csv('data/smartphone.csv')
pressure = smartphone[smartphone['source']=='pressure']
pressure['timestamp'] = pd.to_datetime(pressure['timestamp'])
pressure = pressure.sort_values('timestamp')
pressure = pressure.set_index('timestamp')

pressure['pressure'] = pressure['values'].map(lambda hr: float(hr.split('\'')[1]))
pressure = pressure.groupby(pd.TimeGrouper(freq='Min')).agg({
    'pressure': 'mean',
})
```

```python
smartphone_acceleration = smartphone[smartphone['source']=='linear_acceleration']

smartphone_acceleration['timestamp'] = pd.to_datetime(smartphone_acceleration['timestamp'])
smartphone_acceleration = smartphone_acceleration.sort_values('timestamp')
smartphone_acceleration = smartphone_acceleration.set_index('timestamp')

smartphone_acceleration['smartphone_acceleration'] = smartphone_acceleration['values'].map(
    lambda hr: np.sqrt(
        float(hr.split('\'')[1]) ** 2 + 
        float(hr.split('\'')[3]) ** 2 + 
        float(hr.split('\'')[5]) ** 2
    )
)
smartphone_acceleration = smartphone_acceleration.groupby(pd.TimeGrouper(freq='Min')).agg({
    'smartphone_acceleration': 'mean',
})
```

```python
report = pd.read_csv('data/report.csv')
labels = pd.DataFrame(columns=['timestamp', 'activity'])
for i, r in report.iterrows():
    min_diff = int(
        (
            datetime.datetime.strptime(r['to'], '%m/%d/%Y %H:%M') - 
             datetime.datetime.strptime(r['from'], '%m/%d/%Y %H:%M')
        ).total_seconds() / 60
    )
    for i in range(min_diff):
        labels = labels.append(pd.DataFrame(
            [{
                'timestamp': datetime.datetime.strptime(r['from'], '%m/%d/%Y %H:%M') + datetime.timedelta(minutes=i),
                'activity': r['activity_type']
            }]
        ), ignore_index=True)
labels = labels.set_index('timestamp')
```

```python
agg = (eog
        .join(pressure)
        .join(heart_rate)
        .join(smartphone_acceleration)
        .join(smartwatch_acceleration)
        .join(labels)
      )
agg
```

```python
agg.to_csv('data/agg.csv', index_label='timestamp')
```
